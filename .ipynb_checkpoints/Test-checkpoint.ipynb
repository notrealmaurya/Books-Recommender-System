```python
ratings= pd.read_csv("DS_Book_Ratings.csv", sep=";", on_bad_lines="skip", encoding="latin-1")
```

```python
ratings.head(2)
```

```python
ratings.rename(columns={"User-ID":"uid","Book-Rating":"rating"
                     },inplace=True)
```

```python
ratings['uid'].value_counts()
```

```python
x=ratings['uid'].value_counts() > 200
x[x].shape
```

```python
y= x[x].index
y
```

```python
ratings=ratings[ratings['uid'].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
books.head(2)
```

```python
rating_with_books=ratings.merge(books,on="ISBN")
```

```python
rating_with_books.head()
```

```python
rating_with_books.shape
```

```python
num_rating = rating_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.shape
```

```python
num_rating.head()
```

```python
num_rating.rename(columns={"rating":"num_of_rating"},inplace=True)
```

```python
num_rating.head()
```

```python
final_rating=rating_with_books.merge(num_rating, on='title')
```

```python
final_rating.head(2
                )
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['num_of_rating']>=50]
```

```python
final_rating.shape
```

```python
final_rating.sample(10)
```

```python
final_rating.drop_duplicates(['uid','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot_table=final_rating.pivot_table(columns='uid',index='title',values='rating')
```

```python
book_pivot_table
```

```python
book_pivot_table.shape
```

```python
book_pivot_table.fillna(0,inplace=True)
```

```python
book_pivot_table
```

```python
from scipy.sparse import  csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot_table)
```

```python
book_sparse.shape
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot_table.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot_table.index[suggestion[i]])
```

```python
import pickle 
pickle.dump(model,open('artifacts/model.pkl','wb'))
pickle.dump(books_name,open('artifacts/book_names.pkl','wb'))
pickle.dump(final_rating,open('artifacts/final_rating.pkl','wb'))
pickle.dump(book_pivot_table,open('artifacts/book_pivot_table.pkl','wb'))
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot_table.index==book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot_table.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestion)):
        books = book_pivot_table.index[suggestion[i]]
        for j in books: 
            print(j)
```
